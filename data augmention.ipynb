```python
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import array_to_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import load_img
```

```python
import cv2 as cv
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
from random import*
from numpy.random import randn
from numpy.random import randint
from google.colab import drive
drive.mount('/content/gdrive')
```

```python

   
# Initialising the ImageDataGenerator class.
# We will pass in the augmentation parameters in the constructor.
datagen = ImageDataGenerator(
        rotation_range = 40,
        shear_range = 0.2,
        zoom_range = 0.2,
        horizontal_flip = True,
        brightness_range = (0.5, 1.5))
path =os.path.join("./gdrive/My Drive/newdata/zaha hadid building")
data=[]
for filename in os.listdir(path):
        img = cv.imread(os.path.join(path, filename))    
        # Loading a sample image  
        # Converting the input sample image to an array
        x = img_to_array(img)
        # Reshaping the input image
        x = x.reshape((1, ) + x.shape) 
        i = 0
        for batch in datagen.flow(x, batch_size = 1,
                                  save_to_dir =os.path.join("./gdrive/My Drive/newdata/zaha hadid building"), 
                                  save_prefix ='image', save_format ='jpeg'):
            i += 1
            if i > 5:
                break
```

```python
import Augmentor
# Passing the path of the image directory
p = Augmentor.Pipeline(r"C:\Users\ASUS\Desktop\Google-Image-Scraper-master\Google-Image-Scraper-master\photos\zaha hadid building")
  
# Defining augmentation parameters and generating 5 samples
p.flip_left_right(0.5)
p.black_and_white(0.1)
p.rotate(0.3, 10, 10)
p.skew(0.4, 0.5)
p.zoom(probability = 0.2, min_factor = 1.1, max_factor = 1.5)
p.sample(5)
```
